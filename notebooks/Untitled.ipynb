```python
from torch.utils.data import Dataset, DataLoader
import torch
class MyDataset(Dataset):
    def __init__(self):
        self.X = [0, 10, 20, 30, 40, 50, 60, 70]
        self.y = [0, 1, 2, 3, 4, 5, 6, 7]
        self.num_data = 8

    def __len__(self):
        return self.num_data

    def __getitem__(self, i):
        x = self.X[i]
        y = self.y[i]
        return x, y
dataset = MyDataset()
loader1 = DataLoader(dataset, batch_size=2, shuffle=True)
loader2 = DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
torch.manual_seed(0)
```

```python
import json
from nltk.tokenize import word_tokenize
import gensim

with open('data/squad1.1/train-v1.1.json', encoding='utf-8') as f:
    data_dict = json.load(f)
# w2v_model = gensim.models.KeyedVectors.load_word2vec_format(
#     './data/GoogleNews-vectors-negative300.bin', binary=True
# )
w2v_model = gensim.models.KeyedVectors.load_word2vec_format(
    './data/GoogleNews-vectors-negative300-hard-debiased.bin', binary=True
)
```

```python
def isSign(word):
    return len(word) == 1 and not word in w2v_model.key_to_index
```

```python
num_context = 0
num_question = 0
max_words_context = 0
max_words_question = 0
set1 = set()
set2 = set()
for title in data_dict['data']:
    for paragraph in title['paragraphs']:
        num_context += 1
        context = word_tokenize(paragraph['context'].lower())
        for word in context:
            if isSign(word):
                set1.add(word)
        for qa in paragraph['qas']:
            num_question += 1
            question = word_tokenize(qa['question'].lower())
            for word in question:
                if isSign(word):
                    set2.add(word)
```

```python
num_context = 0
num_question = 0
max_words_context = 0
max_words_question = 0
for title in data_dict['data']:
    for paragraph in title['paragraphs']:
        num_context += 1
        num_words_context = len([
            word for word in word_tokenize(paragraph['context'].lower())
            if not isSign(word)
        ])
        if num_words_context > max_words_context:
            max_words_context = num_words_context
        for qa in paragraph['qas']:
            num_question += 1
            num_words_question = len([
                word for word in word_tokenize(qa['question'].lower())
                if not isSign(word)
            ])
            if num_words_question > max_words_question:
                max_words_question = num_words_question
```
